## STEP 10 - Clean Entore Dataset

Now we will apply all the changes to the patients list.

1. Eliminate all the extra spaces, as the same with the underscore.
2. Turn all the ages to whole numbers
3. Separate all the names and lastnames in a sublist

Save the modified list as a new variant called "users_clean" and show it on the screen

```python
patients_clean = []

for patient in patients:
    try:
        patient_id = int(patient[0])
    except ValueError:
        print("The patient's ID must be a number")
    patient_name = patient[1].strip().replace("_", " ").split()
    patient_age = int(patient[2])
    services = [c.lower() for c in patient[3]]
    expenses = patient[4]
    total_expenses = sum(expenses)
    patients_clean.append([patient_id, patient_name, patient_age, services, expenses, total_expenses])

for patient in patients_clean:
    print(f'ID: {patient[0]} | Patient Name: {patient[1]} | Patient age: {patient[2]} | Services: {patient[3]} | Expenses: {patient[4]} | Total expenses: {patient[5]}')
```
